```python
import pandas as pd
import re
```

```python
#reading the csv file
df = pd.read_csv('grpcal2.csv')
```

```python
column_to_move = 'UEI'
new_column_order = [column_to_move] + [col for col in df.columns if col != column_to_move]
df = df[new_column_order]
```

```python
# Function to remove duplicates from a cell
def remove_duplicates(cell):
    items = cell.split(", ")  # Split the string into a list based on comma and space
    unique_items = list(dict.fromkeys(items))  # Remove duplicates and preserve order
    return ", ".join(unique_items)  # Join the items back into a single string
# Apply the function to each cell in the 'Certs' column
df['Certifications'] = df['Certifications'].apply(remove_duplicates)
df
```

```python
def format_phone_number(phone):
    if pd.isna(phone):
        return None
    phone = str(phone)
    cleaned_phone = re.sub(r"[^\d]", "", phone)
    if len(cleaned_phone) == 10:
        formatted_phone = f"{cleaned_phone[:3]}-{cleaned_phone[3:6]}-{cleaned_phone[6:]}"
        return formatted_phone
    else:
        return None
df['Office Phone Number'] = df['Office Phone Number'].apply(format_phone_number)
df
```
